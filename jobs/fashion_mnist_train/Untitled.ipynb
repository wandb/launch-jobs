```python
import pandas as pd
import folium

# Assume we have a DataFrame df with columns 'lat', 'lon', 'metric_a', 'metric_b'
df = pd.DataFrame({
    'lat': [37.7749, 37.8715, 37.7749],  # Sample latitudes
    'lon': [-122.4194, -122.2730, -122.4194],  # Sample longitudes
    'metric_a': [1, 2, 3],  # Sample metric A values
    'metric_b': [10, 20, 30]  # Sample metric B values
})

# Create a map centered around the mean latitude and longitude values
m = folium.Map(location=[df['lat'].mean(), df['lon'].mean()], zoom_start=13)

# Add circles to the map
for i in range(0,len(df)):
    folium.Circle(
        location=[df.iloc[i]['lat'], df.iloc[i]['lon']],
        popup='Metric A: ' + str(df.iloc[i]['metric_a']) + ', Metric B: ' + str(df.iloc[i]['metric_b']),
        radius=float(df.iloc[i]['metric_b'])*10,
        color='crimson',
        fill=True,
        fill_color='crimson'
    ).add_to(m)

# Display the map
m
```
